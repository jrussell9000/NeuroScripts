```python
import os,sys,nipype
import nipype.pipeline as pe
import nipype.interfaces.fsl as fsl
from pathlib import Path

def fsl_tbss(subjdir, sesdir):
    sourcedir = Path(sesdir,'dti')
    fslprocdir = Path(sourcedir, 'fsl')
    fslprocdir.mkdir(exist_ok=True)
    tbssdir = Path(fslprocdir, 'tbss')
    tbssdir.mkdir(exist_ok=True)
    subjroot = "_".join([subjdir.name, sesdir.name])
    sourcedwi = sourcedir / (subjroot + "_ppd.mif")
    sourcedwimask = sourcedir / (subjroot + "_mask_ppd.mif")
    sourcebvec = fslprocdir / (subjroot + ".bvec")
    sourcebval = fslprocdir / (subjroot + ".bval")
    sourcedwiniigz = fslprocdir / (subjroot + "_ppd.nii.gz")
    sourcedwimaskniigz = fslprocdir / (subjroot + "_mask_ppd.nii.gz")
    tbssbasename = Path(tbssdir, subjroot)
    !mrconvert -force -export_grad_fsl $sourcebvec $sourcebval $sourcedwi $sourcedwiniigz                                
    !mrconvert -force $sourcedwimask $sourcedwimaskniigz
    !dtifit -k $sourcedwiniigz -o $tbssbasename -m $sourcedwimaskniigz -r $sourcebvec -b $sourcebval -V
```

```python
bidsproc_dir = Path("/Users/jdrussell3/youthptsd/BIDS_Processed")
subjdirs = (subjdir for subjdir in bidsproc_dir.iterdir() if subjdir.is_dir() and 'sub' in subjdir.name)

for subjdir in sorted(subjdirs):
    sesdirs = (sesdir for sesdir in subjdir.iterdir() if sesdir.is_dir())
    for sesdir in sorted(sesdirs):
        dtidir = sesdir/'dti'
        if dtidir.exists():
            try:
                fsl_tbss(subjdir, sesdir)
            except Exception as e:
                print(e)
                errfile = Path(bidsproc_dir / 'tbss_error_log.txt')
                with open(errfile, 'a+') as errorfile:
                    errorfile.write(str('Error processing ' + subjdir.name + ', ' + \
                                        sesdir.name + ':\n' + str(e) + '\n'))
                next
```
